```
import pandas as pd
df_final= pd.read_csv('/work/FINAL_draft.tsv.gz',sep='\t')
df_final.head()
```

```
#supprimer les lignes ou genre == '\N'
df_final = df_final[df_final['genres'] != '\\N']
df_final.shape
```

```
#supprimer les lignes ou release_date =='inconnu'
df_final=df_final[df_final['release_date']!='inconnu']
df_final
```

```
# Convertir release_date en datetime
df_final['release_date'] = pd.to_datetime(df_final['release_date'])

# Créer les nouvelles colonnes pour le mois et l'année
df_final['mois de sortie'] = df_final['release_date'].dt.month
df_final['annee de sortie'] = df_final['release_date'].dt.year

# Afficher les statistiques descriptives de la colonne 'annee de sortie'
df_final['annee de sortie'].describe()
```

```
df_final.head()
```

```
import string

def remove_punctuations(text):
    for punctuation in string.punctuation:
        text = text.replace(punctuation, '')
    return text


df_final["societe_prod"] = df_final['societe_prod'].apply(remove_punctuations)
df_final["titre_intl"] = df_final['titre_INTL'].apply(remove_punctuations)
df_final["societe_prod"] = df_final['societe_prod'].apply(remove_punctuations)

df_final.head(40)
```

```
# Remplacer les lignes vides de societe_prod par 'inconnu'
df_final['societe_prod'] = df_final['societe_prod'].fillna('inconnu')
```

```
df_final.drop(columns=['release_date'],axis=1,inplace=True)
df_final.drop(columns=['titre_INTL'],axis=1,inplace=True)
df_final
```

```
df_final_renamed = df_final.rename(columns={
    'primaryName': 'prenom_nom',
    'category': 'role',
    'nconst': 'personneID',  # Added missing comma here
    'langue_VO': 'pays'
})
df_final_renamed.head(40)
```
